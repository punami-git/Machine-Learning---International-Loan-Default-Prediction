```python
#load our libraries
import pandas as pd
import numpy as np
```

```python
#load our educational data
df1 = pd.read_excel('All_Countries_Edu_Data.xlsx')
```

```python
df1
```

```python
#load the employment data
df2 = pd.read_excel('Employment-data-final-all-countires.xlsx')
```

```python
df2
```

```python
#load in the socioeconomic data
df3 = pd.read_excel('DataForTable2.1.xlsx')
```

```python
df3
```

```python
#load in the health data
df4 = pd.read_csv('pivoted_data.csv')
```

```python
df4 
```

```python
#merge dataframe one and two
combined_df2 = pd.merge(df1, df2, how='outer', on=['Country', 'Year'])
```

```python
combined_df2
```

```python
#add dataframe 3
combined_df3 = pd.merge(combined_df2, df3, how='outer', on=['Country', 'Year'])
```

```python
#add dataframe 4
combined_df4 = pd.merge(combined_df3, df4, how='outer', on=['Country', 'Year'])
```

```python
combined_df4
```

```python
combined_df4['Country'].unique()
```

```python
#save our combined dataframe as a csv
output_file_path = 'Only_X_Columns.xlsx'

#pivoted_df.to_csv(output_file_path, index=True)
combined_df4.to_excel(output_file_path, index=True)

# Displaying the path to the new CSV file
output_file_path
```

```python
#read in our target variable, debt percent of gdp
df5 = pd.read_csv('sorted_debt_data.csv')
df5
```

```python
#add it to our 4 combined dataframes
combined_df5 = pd.merge(df5, combined_df4, how='left', on=['Country', 'Year'])
```

```python
combined_df5
```

```python
#this is our final data that contains all our feature variables, and our target variable.
output_file_path = 'FFFfffinal_Data.xlsx'

#pivoted_df.to_csv(output_file_path, index=True)
combined_df5.to_excel(output_file_path, index=True)

# Displaying the path to the new CSV file
output_file_path
```
